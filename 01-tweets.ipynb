```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
from sqlalchemy import func

from frontline.db import session
from frontline.models import Tweet
```

# Total tweets

```python
Tweet.query.count()
```

# Number of unique users

```python
session.query(func.distinct(Tweet.actor_id)).count()
```

# Tweets per day

```python
posted = [r.date() for r, in session.query(Tweet.posted_time)]
```

```python
plt.figure(figsize=(10, 6))
plt.title('Tweets per day (Powertrack)')
plt.xlabel('Date')
plt.ylabel('Number of Tweets')
_ = plt.hist(posted, bins=len(set(posted)), linewidth=1, edgecolor='black')
```
